[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
def sum_numbers(*input_numbers):
    total = 0
    for num in input_numbers:
        total += num
    return total

result = sum_numbers(1, 10, 100, 1000)
print(result)
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    for i in range(n+1):
        if i % 2 == 1:
            yield i


# for 루프를 사용하여 전체 순회
for num in odd_numbers(10):
    print(num)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max():
    numbers = []

    # 사용자로부터 3개의 숫자 입력받아 리스트로 저장
    for i in range(3):
        num = int(input(f"{i+1}번째 숫자를 입력하세요: "))
        numbers.append(num)

    # 리스트를 튜플로 변환
    numbers_tuple = tuple(numbers)

    min_value = min(numbers_tuple)
    max_value = max(numbers_tuple)

    # 결과 출력
    print("\n----------------------------------")
    print(f"입력된 숫자: {numbers_tuple}")
    print(f"최소값: {min_value}")
    print(f"최대값: {max_value}")
    print("----------------------------------")

# 함수 실행
min_max()
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False

result = is_even(11)
print(result)
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    facto_num = 1
    for i in range(1,n+1):
        facto_num *= i
    return facto_num

result = factorial(5)
print(result)
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(str1, str2):
    if len(str1) > len(str2):
        return str1
    elif len(str1) < len(str2):
        return str2
    else:
        return str1

str1 = "Helloooo"
str2 = "World"
result = longer_string(str1, str2)
print(result)
```
