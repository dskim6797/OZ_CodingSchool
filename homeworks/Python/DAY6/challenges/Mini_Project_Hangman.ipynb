##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
# 행맨 그림
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |  ㅡ|
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |  ㅡ|ㅡ
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |  ㅡ|ㅡ
     |    |
     |   |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |  ㅡ|ㅡ
     |    |
     |   | |
     |
    ---""",
]
```

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]


class HangmanGame:
    def __init__(self):
        self.secret_word = random.choice(words)
        self.guessed_letters = []
        self.missed_letters = set()               # 틀린 글자 집합
        self.correct_letters = set()              # 맞춘 글자 집합

        self.missed_count = 0                     # 틀린 횟수
        self.max_misses = len(hangman_pics) - 1   # 최대 허용 틀린 횟수 (6)

    def display(self, missed_count: int):
        # 틀린 횟수를 기반으로 행맨 그림을 출력
        if 0 <= missed_count < len(hangman_pics):
            print("\n" + hangman_pics[missed_count])
        else:
            print("틀린 횟수 최대치 초과")
            print("\n" + hangman_pics[-1])

        # 단어 진행 상황 출력
        blanks = ''
        for letter in self.secret_word:
            if letter in self.correct_letters:
                blanks += letter + ' '
            else:
                blanks += '_ '

        print(f"\n단어: {blanks}")
        print(f"틀린 글자 ({self.missed_count}/{self.max_misses}): {sorted(list(self.missed_letters))}")
        print("-" * 30)

    def play(self):
        print("======== 행맨 게임 시작 ========")
        game_over = False

        while not game_over:
            self.display(self.missed_count)

            # 사용자 입력 받기
            guess = input("글자 하나를 추측해보세요: ").lower()

            # 입력값 유효성 검사
            if not guess.isalpha() or len(guess) != 1:
                print("⚠️ 오류: 알파벳 한 글자만 입력해주세요.")
                continue

            # 이미 시도한 글자인지 확인
            if guess in self.missed_letters or guess in self.correct_letters:
                print(f"⚠️ '{guess}'는 이미 시도한 글자입니다. 다른 글자를 입력해주세요.")
                continue

            # 정답/오답 처리
            if guess in self.secret_word:
                self.correct_letters.add(guess)
                print(f"✅ 정답입니다! '{guess}' 글자를 맞췄습니다.")
            else:
                self.missed_letters.add(guess)
                self.missed_count += 1
                print(f"❌ 오답입니다! '{guess}' 글자가 없습니다. 남은 목숨: {self.max_misses-self.missed_count}")

            # 게임 종료 조건 확인
            # 승리
            # 모든 글자를 다 맞췄는지 확인
            # secret_word의 모든 고유 글자가 correct_letters에 포함되어 있는지 확인
            if all(letter in self.correct_letters for letter in set(self.secret_word)):
                print(f"\n🎉 축하합니다! 정답은 '{self.secret_word}'입니다. 당신이 이겼습니다!")
                game_over = True

            # 패배
            # 틀린 횟수가 최대 허용 횟수에 도달했는지 확인
            elif self.missed_count >= self.max_misses:
                self.display(self.missed_count) # 마지막 그림 출력
                print(f"\n😭 게임 오버! 정답은 '{self.secret_word}'였습니다. 행맨이 완성되었습니다.")
                game_over = True

if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
