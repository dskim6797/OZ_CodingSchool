__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

max_value = max(nums)  # 최댓값
min_value = min(nums)  # 최솟값
average_value = sum(nums) / len(nums)  # 평균

print("최댓값:", max_value)
print("최솟값:", min_value)
print("평균:", average_value)
````

__[문제7]__

아래 순서대로 주어진 수강생 이름과 등급을 하나의 딕셔너리로 만드는 코드를 작성해주세요


names : 나이썬, 김자바, 김씨, 박씨샵  
grades :   B       B      A      C


````python
names = ['나이썬', '김자바', '김씨', '박씨샵']
grades_str = 'B B A C'
grades = grades_str.split() # ['B', 'B', 'A', 'C']

student_info = dict(zip(names, grades))

print("생성된 수강생 딕셔너리:")
print(student_info)
````
