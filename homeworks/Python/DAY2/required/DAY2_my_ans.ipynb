## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다.

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

### [동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

<br>

* 변수명을 영어로 변경해서 풀어보세요 !

```python
wizard_STR = 10
wizard_DEX = 20
wizard_INT = 80
wizard_STMN = 50

warrior_STR = 80
warrior_DEX = 30
warrior_INT = 20
warrior_STMN = 100

rogue_STR = 40
rogue_DEX = 80
rogue_INT = 30
rogue_STMN = 60

healer_STR = 20
healer_DEX = 40
healer_INT = 70
healer_STMN =
```

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
if wizard_INT >= 50:
    print("마법을 사용할 수 있습니다.")
else:
    print("지능이 부족하여 마법을 사용할 수 없습니다.")
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
if warrior_STR >= 60 and warrior_STMN >= 80:
    print("자신의 힘을 사용하여 강력한 공격을 합니다.")
else:
    print("힘 또는 체력이 부족합니다.")
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
if rogue_DEX >= 70 or rogue_STR >= 50:
    print(" 몬스터를 회피하고 공격합니다.")
else:
    print("민첩과 힘이 부족합니다.")
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
if healer_INT >= 60 and healer_STMN >= 50:
    print("파티의 체력을 회복시키기 위한 마법을 사용합니다.")
else:
    print("지능 또는 체력이 부족합니다.")
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
wizard_EXP = 120
warrior_EXP = 150
rogue_EXP = 110
healer_EXP = 90

wizard_levelup = wizard_EXP>=100
warrior_levelup = warrior_EXP>=100
rogue_levelup = rogue_EXP>=100
healer_levelup = healer_EXP>=100

if wizard_levelup:
    print("마법사 레벨업!")
else:
    print("마법사가 레벨업하지 못했습니다. 경험치를 더 모으세요.")
if warrior_levelup:
    print("전사 레벨업!")
else:
    print("전사가 레벨업하지 못했습니다. 경험치를 더 모으세요.")
if rogue_levelup:
    print("도적 레벨업!")
else:
    print("도적이 레벨업하지 못했습니다. 경험치를 더 모으세요.")
if healer_levelup:
    print("힐러 레벨업!")
else:
    print("힐러가 레벨업하지 못했습니다. 경험치를 더 모으세요.")
```
