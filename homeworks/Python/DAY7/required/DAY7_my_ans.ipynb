[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요.

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
print("숫자를 입력하세요. 입력을 종료하고 싶으면 'done'을 입력하세요.")
numbers = []

while True:
    input_num = input(f"{len(numbers)+1}번째 숫자 입력:")

    if input_num == 'done':
        print("입력을 종료합니다.")
        break

    try:
        number = int(input_num)
        numbers.append(number)

    except ValueError:
        print(f"잘못된 값: {input_num}. 숫자를 입력해주세요.")

print(numbers)
````

[문제 4]

사용자 정의 예외 NegativeNumberError를 생성하고, 사용자로부터 숫자를 입력받아 음수일 경우 NegativeNumberError를 발생시키는 프로그램을 작성하세요.

- 예외 처리 구문을 사용하여 음수 입력 시 "양수를 입력해주세요."라는 메시지를 출력

````python
class NegativeNumberError(Exception):
    def __init__(self) -> None:
        self.contact_name = "NegativeNumberError"
        self.message = "음수가 입력되었습니다. 양수를 입력해주세요."
        super().__init__()

    def __str__(self) -> str:
        return f"{self.contact_name}: {self.message}"

try:
    number = int(input("양수를 입력하세요: "))
    if number < 0:
        raise NegativeNumberError()

except NegativeNumberError as e:
    print(e)

except ValueError: # 숫자 외 input값에 대한 예외처리
    print("숫자 형식이 잘못되었습니다.")
````

[문제 5]

datetime 모듈을 사용하여 현재 날짜와 시간을 출력하는 프로그램을 작성하세요.

또한, 사용자로부터 태어난 년도를 입력받아 현재 나이를 계산하여 출력하세요.

````python
import datetime
import pytz

def cal_user_age(birth_year, current_year):
    user_age = current_year-birth_year
    return user_age


SEOUL_TZ = pytz.timezone('Asia/Seoul') # 서울 시간대 정의
current_time = datetime.datetime.now(SEOUL_TZ) # 서울 시간대 기준으로 현재시간 가져오기
formatted_time = current_time.strftime("%Y-%m-%d %H:%M:%S") # (YYYY-MM-DD HH:MM:SS)형식으로 포맷팅
print(f"현재 날짜와 시간을 출력합니다.\n{formatted_time}\n")

my_birth_year = int(input("사용자의 년도를 입력하면 현재 나이를 계산하여 출력합니다.\n태어난 년도: "))
my_age = cal_user_age(my_birth_year,current_time.year)
print(f"{my_birth_year}년생은 {my_age}살 입니다.")
````

[문제 6]

requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

````python
import requests

# 스크래핑 할 웹 URL
URL = "https://www.naver.com/" # kream.co.kr get 요청시 상태코드 500에러 출력. naver.com으로 URL 수정했습니다.

# 세션 객체 생성
session = requests.Session()
session.headers.update({
    'User-Agnet': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Whale/4.33.325.17 Safari/537.36',
}) # 클라이언트 헤더

# HTTP GET 요청 보내기
try:
    response = session.get(URL)

    # 응답 상태 확인
    if response.status_code == 200:
        (f"요청 성공. 상태 코드: {response.status_code}")

        # HTML 가져오기
        print(f"{URL}에서 HTML 데이터를 가져옵니다.(첫 100자까지 출력 합니다.)")
        print("-" * 50)
        html = response.text
        print(html[:100])
        print("-" * 50)

    else:
        print(f"요청 실패. 상태 코드: {response.status_code}")

# HTTP GET 요청 오류 (네트워크 오류, DNS 오류 등)
except requests.exceptions.RequestException as e:
    print(f"네트워크 요청 중 오류 발생: {e}")
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

# 임의의 함수 실행 시간을 측정하여 출력하는 데코레이터
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time() # 함수 실행 전 시간
        result = func(*args, **kwargs) # 함수 실행
        end_time = time.time() # 함수 실행 후 시간

        # 함수 실행 시간 계산 및 출력
        execution_time = end_time - start_time
        print(f"함수 '{func.__name__}'의 실행 시간: {execution_time:.2f}초")

        # 실행 된 함수의 반환 값 반환
        return result

    return wrapper

@timer_decorator
def example_slow_function(n):
    print(f"1부터 {n}까지 합계를 계산합니다...")
    total = 0
    for i in range(n):
        total += i
    time.sleep(0.05) # 0.5초 대기 추가
    return total

# 함수 실행
result = example_slow_function(1000000)
print(f"함수 반환 값: {result}")
````
